```python
import os
import subprocess

def convert_heic_to_png(folder):
    print(f'Converting in {folder}')
    files = [f for f in os.listdir(folder) if f.lower().endswith('.heic')]
    converted = []
    for file in sorted(files):
        src = os.path.join(folder, file)
        dst = os.path.join(folder, os.path.splitext(file)[0] + '.png')
        if os.path.exists(dst):
            converted.append(dst)
            continue
        result = subprocess.run(['magick', src, dst], capture_output=True, text=True)
        if result.returncode != 0:
            raise RuntimeError(f'Conversion failed for {src}: {result.stderr}')
        converted.append(dst)
    return converted

base = '/Users/madhumitamoi/Madhumita-Moi-Portfolio/travel_photography'
aml = os.path.join(base, 'amalfi coast')
swi = os.path.join(base, 'switzerland')

converted_amalfi = convert_heic_to_png(aml)
converted_switzerland = convert_heic_to_png(swi)

print('Amalfi converted:', len(converted_amalfi))
print('Switzerland converted:', len(converted_switzerland))
```
